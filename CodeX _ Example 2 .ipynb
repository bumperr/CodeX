Preparing X and Y values


```python
# Putting feature variable to X
X = advertising_dataset['TV']
## Reshape 'X' to a 2D array
X = X.values.reshape(-1, 1)

# Putting response variable to y
y = advertising_dataset['Sales']
## Reshape 'y' to a 2D array
y= y.values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Equation of linear regression
y=c+m1x1+m2x2+...+mnxn

y
  is the response
c
  is the intercept
m1
  is the coefficient for the first feature
mn
  is the coefficient for the nth feature
In our case:

y=c+m1Ã—TV

The  m
  values are called the model coefficients or model parameters.

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train) # Training the algorithm
```

```python
c = round(regressor.intercept_[0], 2)
m = round(regressor.coef_[0][0], 2)

# Format the equation with two decimal places
equation = f"y = {m}x + {c}"

print(equation)
```

```python
y_pred=regressor.predict(x_test)
df_results = pd.DataFrame({
    'X_test': x_test.flatten(),
    'Actual_Y': y_test.flatten(),
    'Predicted_Y': y_pred.flatten()
})
print(df_results.head(10))
```

```python
from matplotlib import pyplot as plt
plt.plot(x_train,regressor.predict(x_train), color='red')
plt.scatter(X,y)
plt.title('Tv vs Sales')
plt.xlabel("TV")
plt.ylabel('Sales ')
plt.show()
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
print(f"MAE:{mae} ,\n MSE:{mse} ,\n RMSE:{rmse},\n R2:{r2}")
```

```python
train_score_lr = regressor.score(x_train, y_train)
test_score_lr = regressor.score(x_test, y_test)

print("The train score for lr model is: ", train_score_lr)
print("The test score for lr model is: ",test_score_lr)

```
